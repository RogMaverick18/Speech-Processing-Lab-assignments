```python
!pip install soundfile matplotlib numpy

import soundfile as sf
import os
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio, display

print("="*40)
print("OBJECTIVE 1: USER VOICE ANALYSIS")
print("="*40)

USER_VOICE_PATH = "D:\\Amrita Vishwa Vidyapeetham\\Semester 6\\Speech\\Lab\\Recording.wav"

if os.path.exists(USER_VOICE_PATH):
    file_info = sf.info(USER_VOICE_PATH)

    print(f"File Name     : {USER_VOICE_PATH}")
    print(f"Sampling Rate : {file_info.samplerate} Hz")
    print(f"Bit Depth     : {file_info.subtype}")
    print(f"Duration      : {file_info.duration:.2f} seconds")
    print(f"Channels      : {file_info.channels}")
    print("-" * 40)

    print("Playing Audio:")
    display(Audio(filename=USER_VOICE_PATH))

    print("Plotting Waveform...")
    data, samplerate = sf.read(USER_VOICE_PATH)
    time = np.linspace(0, len(data) / samplerate, num=len(data))

    plt.figure(figsize=(14, 5))
    if len(data.shape) > 1:
        plt.plot(time, data[:, 0], label="Left Channel", alpha=0.7)
        plt.plot(time, data[:, 1], label="Right Channel", alpha=0.7)
        plt.legend()
    else:
        plt.plot(time, data)

    plt.title(f"Waveform of Recorded Audio: {os.path.basename(USER_VOICE_PATH)}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.grid(True, alpha=0.3)
    plt.show()

else:
    print(f"[ERROR] File not found at: {USER_VOICE_PATH}")
    print("Please upload your .wav file and update the USER_VOICE_PATH variable.")
```
